## 1-3. tick & ticklabels
- Îã§Î•∏ Ï¢ÖÎ•òÏùò plotÍ≥ºÎäî Îã¨Î¶¨ ax.bar Î©îÏÑúÎìú ÏïàÏóê Îß§Í∞úÎ≥ÄÏàòÎ°ú Ï†ÑÎã¨ÌïúÎã§.

```python
np.random.seed(0)

n_data = 10
data = np.random.uniform(10, 100, (n_data, ))
data_idx = np.arange(n_data)
background = 100*np.ones(n_data)

data_labels = ['class ' + str(i) for i in range(n_data)]

fig, ax = plt.subplots(figsize=(10, 7))
ax.bar(data_idx, data)

major_yticks = np.arange(0, 101, 20)
major_yticklabels = [str(p) + '%' for p in major_yticks]
minor_yticks = np.arange(0, 101, 5)

ax.set_xticks(data_idx)
ax.set_xticklabels(data_labels, rotation=-30, ha='left')
ax.set_yticks(major_yticks)
ax.set_yticklabels(major_yticklabels)
ax.set_yticks(minor_yticks, minor=True)

ax.grid(axis='y', which='major')
ax.grid(axis='y', which='minor', linestyle=':')

ax.tick_params(labelsize=15)
```

## 1-4. Ï†ïÎ†¨
üî∏ argsortÍ∞Ä ÌïµÏã¨

```python
np.random.seed(0)

n_data = 10
data_idx = np.arange(n_data)
data = np.random.uniform(0, 10, n_data)  # ÌïµÏã¨
data_labels = np.array(['class ' + str(i+1) for i in range(n_data)])

# Ï†ïÎ†¨
sort_idx = np.argsort(data)
sort_idx_rev = np.argsort(data)[::-1]
data_sort = data[sort_idx]
data_sort_rev = data[sort_idx_rev]
data_labels_sort = data_labels[sort_idx]
data_labels_sort_rev = data_labels[sort_idx_rev]

fig, axes = plt.subplots(3, 1, figsize=(10, 12))
axes[0].bar(data_idx, data, tick_label=data_labels)
axes[1].bar(data_idx, data_sort, tick_label=data_labels_sort)
axes[2].bar(data_idx, data_sort_rev, tick_label=data_labels_sort_rev);
```

## 2. Multiple Barplots

```python
np.random.seed(0)

N = 2
WIDTH = .8
BAR_WIDTH = WIDTH / N

n_data = 10
data1 = np.random.uniform(10, 100, (n_data, ))
data2 = np.random.uniform(10, 100, (n_data, ))
data_idx = np.arange(n_data)

fig, ax = plt.subplots(figsize=(10, 7))
ax.bar(data_idx - BAR_WIDTH/2, data1,
       width=BAR_WIDTH,
       label='data 1')
ax.bar(data_idx + BAR_WIDTH/2, data2,
       width=BAR_WIDTH,
       label='data 2')

xticks = np.arange(-1, 10, .5)
ax.set_xticks(xticks)
ax.set_xticklabels(xticks, rotation=30)
ax.tick_params(labelsize=10)

ax.grid(axis='x')

ax.legend(loc='lower center',
          bbox_to_anchor=(.5, 1),
          fontsize=15,
          ncol=N);
```

```python
# barh
np.random.seed(0)

n_data = 10
data1 = np.random.uniform(20, 40, (n_data,))
data2 = np.random.uniform(10, 20, (n_data,))
background = 50 * np.ones(n_data)
data_idx = np.arange(n_data)
colors = ['tab:blue', 'tab:red']
labels = ['data 1', 'data 2']

fig, ax = plt.subplots(figsize=(10, 7))
ax.barh(data_idx, background,
        facecolor='whitesmoke',
        edgecolor='silver',
        hatch='/')

rects1 = ax.barh(data_idx, data1,
                 color=colors[0],
                 label=labels[0])
rects2 = ax.barh(data_idx, data2,
                 color=colors[1],
                 label=labels[1])
ax.legend(loc='upper center',
          bbox_to_anchor=(.5, -.05),
          fontsize=20,
          ncol=2)

xticks = ax.get_xticks()
xtick_interval = xticks[1] - xticks[0]
for idx, rect in enumerate(rects1):
    y = rect.get_y()
    width = rect.get_width()
    height = rect.get_height()
    
    ax.text(width + .05*xtick_interval,
            y + .5*height,
            str(round(width)),
            va='center',
            fontsize=10,
            color=colors[0])
for idx, rect in enumerate(rects2):
    y = rect.get_y()
    width = rect.get_width()
    height = rect.get_height()
    
    ax.text(width + .05*xtick_interval,
            y + .5*height,
            str(round(width)),
            va='center',
            fontsize=10,
            color=colors[1])    
```

## 5. Ïó∞ÏäµÎ¨∏Ï†ú

### ‚úÖ Q1

```python
xticklabels = np.array(['White', 'Black', 'Gray', 'Red', 'Green',
                             'Blue', 'Yellow', 'Brown', 'Other'])
xticks = np.arange(len(xticklabels))
data = np.array([47.19, 15.82, 14.32, 7.30, 2.64, 5.64, .84, 5.38, .73])

background = 55*np.ones_like(data)

fig, ax = plt.subplots(figsize=(15, 10))
# background plotting
ax.bar(
    xticks, background,
    color='k',
    edgecolor='k',
    alpha=.05,
    hatch='/'
)
# bar plotting
rects = ax.bar(
    xticks, data, color='cornflowerblue'
)
# tick, ticklabels
ax.set_xticks(xticks)
ax.set_xticklabels(
    xticklabels,
    fontsize=30,
    rotation=45
)
ax.tick_params(
    left=False,
    labelleft=False,
    bottom=False
)
# spines
for spine in ax.spines.values():
    spine.set_visible(False)
    
# annotation
for rect in rects:
    width = rect.get_width()
    height = rect.get_height()
    x = rect.get_x()
    
    ax.text(
        x + .5*width, 56,
        str(height) + '%',
        fontsize=20,
        rotation=-45,
        ha='center',
        va='bottom'
    )
    
fig.tight_layout()
```
