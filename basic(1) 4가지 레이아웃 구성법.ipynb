## 1) plt.figure
- 그래프/이미지를 그리기 위한 전체 영역을 확보한다.
- fig 객체를 인스턴스화 한다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='gray')
```

## 2) fig.add_subplot
- 하나의 그래프/이미지를 위한 영역(ax)을 생성한다.
- 그래프/이미지당 하나의 ax를 갖는 것이다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='gray')

ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)
```

```python
# 그리드 시스템을 바라보는 관점을 다르게 하면 레이아웃에 변화를 줄 수 있다.
fig = plt.figure(figsize=(7, 7), facecolor='gray')

# 2행 2열 그리드 시스템의 첫 번째 ax
ax1 = fig.add_subplot(2, 2, 1)
# 2행 2열 그리드 시스템의 두 번째 ax
ax2 = fig.add_subplot(2, 2, 2)

# 2행 1열 그리드 시스템의 두 번째 ax
ax3 = fig.add_subplot(2, 1, 2)
```

```python
# 또 다른 예제
fig = plt.figure(figsize=(7, 7), facecolor='gray')

ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
# 1행 3열 그리드 시스템의 세 번째 ax
ax3 = fig.add_subplot(133)
ax4 = fig.add_subplot(234)
ax5 = fig.add_subplot(235)
```

## 3) plt.subplots
- plt.subplots() = plt.figure() + fig.add_subplot()
- 각 그래프/이미지를 규칙적으로 그릴 수는 있지만 위와 같이 불규칙적인 레이아웃은 구현할 수 없다.
- axes.flat (다차원 배열을 1차원으로 평탄화한다)

```python
# fig, (ax1, ax2) = plt.subplots(2, 1)
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(7, 7), facecolor='orange')
```

```python
# axes의 타입은 넘파이 배열이다.
print(type(axes))

print(axes[0])
print(axes[1])
```

```python
# 다차원 배열을 1차원으로 평탄화한다 (flat 속성)
fig, axes = plt.subplots(2, 2, figsize=(7, 7), facecolor='orange')

print(axes.shape)
print(axes)

for ax in axes.flat:
    print(ax)
    
for idx, ax in enumerate(axes.flat):
    print(idx, ax)
```

## 4) plt.subplot2grid
- fig.add_subplot(), plt.subplots() 메서드로는 구현하기 복잡한 레이아웃을 구성할 수 있다.
- 매개변수 fig의 값을 전달해야 한다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='lightgreen')

ax1 = plt.subplot2grid((2, 1), (0, 0), fig=fig)
ax2 = plt.subplot2grid((2, 1), (1, 0), fig=fig)
```

```python
# 2차원 레이아웃
fig = plt.figure(figsize=(7, 7), facecolor='lightgreen')

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2, fig=fig)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=3, fig=fig)
ax3 = plt.subplot2grid((3, 3), (2, 0), fig=fig)
ax4 = plt.subplot2grid((3, 3), (2, 1), fig=fig)
ax5 = plt.subplot2grid((3, 3), (2, 2), fig=fig)
```

```python
# fig.add_subplot()과 plt.subplot2grid() 메서드를 사용해서 동일한 결과를 얻을 수 있다.
fig = plt.figure(figsize=(7, 7), facecolor='gray')
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 2, 3)
ax3 = fig.add_subplot(2, 2, 4)

fig2 = plt.figure(figsize=(7, 7), facecolor='lightgreen')
ax4 = plt.subplot2grid((2, 2), (0, 0), colspan=2, fig=fig2)
ax5 = plt.subplot2grid((2, 2), (1, 0), fig=fig2)
ax6 = plt.subplot2grid((2, 2), (1, 1), fig=fig2)
```

## 6) fig.add_axes
- fig.add_axes([left, bottom, width, height])
    - left: 왼쪽으로부터 떨어진 거리
    - bottom: 밑으로부터 떨어진 거리
    - width: (1에 대한 비율값) 가로 길이
    - height: (1에 대한 비율값) 세로 길이
- 중첩 그래프/이미지를 그릴 수 있다.
- auto-scaled 레이아웃을 구현할 수 있다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='skyblue')

rect1 = [0.1, 0.1, 0.5, 0.5]
rect2 = [0.7, 0.1, 0.2, 0.5]
rect3 = [0.1, 0.7, 0.8, 0.2]

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```

```python
# 위 코드를 개선
fig = plt.figure(figsize=(7, 7), facecolor='skyblue')

left, bottom = 0.1, 0.1
spacing = 0.05
width1, height1 = 0.5, 0.5
width2 = 1 - (2*left + spacing + width1)
height2 = 1 - (2*bottom + spacing + height1)

rect1 = [left, bottom , width1, height1]
rect2 = [left, bottom + height1 + spacing, 1 - left*2, height2]
rect3 = [left + width1 + spacing, bottom, width2, height1]

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```

```python
# 다른 스케일
fig = plt.figure(figsize=(7, 7), facecolor='skyblue')

left, bottom = 0.1, 0.1
spacing = 0.05
width1, height1 = 0.3, 0.3
width2 = 1 - (2*left + spacing + width1)
height2 = 1 - (2*bottom + spacing + height1)

rect1 = [left, bottom , width1, height1]
rect2 = [left, bottom + height1 + spacing, 1 - left*2, height2]
rect3 = [left + width1 + spacing, bottom, width2, height1]

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```
