## 2) ax.hist 리턴값
- freqs
- bin_edges
- rects

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
freqs, bin_edges, rects = ax.hist(data)
print('freqs(각 빈도 횟수): ', freqs)
print('bin_edges: ', bin_edges)
# rects에는 bar object 10개가 들어있다. (same as before)

for rect in rects:
    x = rect.get_x()
    width = rect.get_width()
    height = rect.get_height()
    
    ax.text(x + .5*width, height + 2,
            str(height),
            fontsize=10,
            ha='center')

ax.tick_params(labelsize=20,
               length=10,
               width=3)
```

### bin_edges

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
freqs, bin_edges, rects = ax.hist(data)

ax.set_xticks(bin_edges)
ax.set_xticklabels(np.around(bin_edges, 2),
                   rotation=30)

ax.tick_params(labelsize=15,
               length=10,
               width=3)
```

```python
# bin_edges가 아닌 bin_centers으로 설정(꿀팁🍯)
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
freqs, bin_edges, rects = ax.hist(data)

bin_centers = (bin_edges[:-1] + bin_edges[1:])/2

ax.set_xticks(bin_centers)
ax.set_xticklabels(np.around(bin_centers, 2),
                   rotation=30)

ax.tick_params(labelsize=15,
               length=10,
               width=3)
```

```python
# 히스토그램에 대한 추가적인 정보 제공
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
freqs, bin_edges, rects = ax.hist(data)

n_bin = len(freqs)
bin_edges_around = np.around(bin_edges, 2)

m_val, M_val = bin_edges_around[0], bin_edges_around[-1]
bin_interval = np.around(bin_edges[1] - bin_edges[0], 2)

text_loc = ax.get_xlim()[0], ax.get_ylim()[1]
text_template = 'Number of Bins: {}\n' \
    'Min, Max Values: {} / {} \n' \
    'Bin Interval: {}'

ax.text(text_loc[0] + .1, text_loc[1] - 5,
        text_template.format(n_bin,
                             m_val, M_val,
                            bin_interval),
        ha='left', va='top',
        fontsize=10)

ax.tick_params(labelsize=15,
               length=10,
               width=3)
```

### bins
- int/sequence 형태의 인자를 전달할 수 있지만 주로 int 값을 전달한다.
- bins의 개수를 적절하게 설정해야 한다.

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(14, 10))

ax.hist(data,
        facecolor='skyblue',
        edgecolor='darkblue',
        linewidth=3)

ax.hist(data,
        bins=20,
        facecolor='coral',
        edgecolor='red',
        linewidth=3)

ax.tick_params(labelsize=15,
               length=10,
               width=3)
```

### density
- 확률이 아니다

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, axes = plt.subplots(2, 1, figsize=(14, 20))

axes[0].hist(data, bins=20)
axes[1].hist(data, bins=20, density=True)

axes[0].tick_params(labelsize=15)
axes[1].tick_params(labelsize=15)
```

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(14, 10))

# 밀도함수 != 확률
freqs, bin_edges, rects = ax.hist(data, density=True)

bin_interval = bin_edges[1] - bin_edges[0]
print(freqs * bin_interval)
print(np.sum(freqs * bin_interval))

ax.tick_params(labelsize=15)
```

```python
# 확률로 변환
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(14, 10))

freqs, bin_edges, rects = ax.hist(data, density=True)

bin_interval = bin_edges[1] - bin_edges[0]
max_freq = freqs.max()
ax.set_ylim([0, max_freq*bin_interval + .05])

for rect in rects:
    height = rect.get_height()
    height *= bin_interval
    rect.set_height(height)
    
ax.set_ylabel('Probability', fontsize=20)

ax.tick_params(labelsize=15)
```

## 4) multiple histograms

### overlapped histograms
- 투명도를 반드시 설정해야 한다.
- bin_width가 다르다는 문제가 발생하기 때문에 조정이 필요하다.

```python
# bin_width가 다른 경우
np.random.seed(0)

n_data1, n_data2 = 300, 300
data1 = np.random.normal(0, 1, (n_data1))
data2 = np.random.normal(5, 3, (n_data2))

fig, ax = plt.subplots(figsize=(14, 10))

ax.hist(data1,
        bins=30,
        alpha=.5)
ax.hist(data2,
        bins=30,
        alpha=.5)

ax.tick_params(labelsize=15)
```

```python
# 솔루션
np.random.seed(0)

n_data1, n_data2 = 300, 300
data1 = np.random.normal(0, 1, (n_data1))
data2 = np.random.normal(5, 3, (n_data2))
data = np.hstack([data1, data2])
m_val, M_val = data.min(), data.max()

data_loc_bin_edges = np.linspace(m_val, M_val, 31)

fig, ax = plt.subplots(figsize=(14, 10))
ax.hist(data1,
        bins=data_loc_bin_edges,
        alpha=.5)
ax.hist(data2,
        bins=data_loc_bin_edges,
        alpha=.5)

ax.tick_params(labelsize=15)
```

```python
np.random.seed(0)

n_data1, n_data2 = 300, 300
data1 = np.random.normal(0, 1, (n_data1))
data2 = np.random.normal(5, 3, (n_data2))
data = np.hstack([data1, data2])
m_val, M_val = data.min(), data.max()

data_loc_bin_edges = np.linspace(m_val, M_val, 31)
data_center = (data_loc_bin_edges[:-1] + data_loc_bin_edges[1:]) / 2
xticks = data_center[::2]

fig, ax = plt.subplots(figsize=(14, 10))
ax.hist(data1,
        bins=data_loc_bin_edges,
        alpha=.5)
ax.hist(data2,
        bins=data_loc_bin_edges,
        alpha=.5)

ax.set_xticks(xticks)
ax.set_xticklabels(np.around(xticks, 2),
                   rotation=30)
ax.set_xticks(data_center,
              minor=True)
ax.tick_params(labelsize=15)

ax.grid()
ax.grid(which='minor', linestyle=':')
```

### 2D data
- 데이터의 크기가 동일하다는 조건을 충족할 때만 적용이 가능하다.
- np.hstack

#### ax.hist with 2D data

```python
n_class, n_data = 2, 300

# data's shape: (300, 2)
data = np.random.normal(0, 1, (n_data, n_class))

fig, ax = plt.subplots(figsize=(14, 10))
ax.hist(data)

ax.tick_params(labelsize=15)
```

```python
np.random.seed(0)
n_data = 300

data1 = np.random.normal(0, 1, (n_data, 1))
data2 = np.random.normal(-5, 2, (n_data, 1))
data3 = np.random.normal(5, 3, (n_data, 1))

data = np.hstack([data1, data2, data3])

fig, ax = plt.subplots(figsize=(14, 10))
ax.hist(data, bins=30)

ax.tick_params(labelsize=15)
```

#### arguments with 2D data
- 앞서 살펴본 alpha, rwidth, cumulative, density 매개변수 모두 설정이 가능하다.
- color

```python
# color 매개변수(배열을 전달)
# cmap 또한 전달 가능

np.random.seed(0)
n_data = 300

data1 = np.random.normal(0, 1, (n_data, 1))
data2 = np.random.normal(-5, 2, (n_data, 1))
data3 = np.random.normal(5, 3, (n_data, 1))

data = np.hstack([data1, data2, data3])

fig, ax = plt.subplots(figsize=(14, 10))
ax.hist(data,
        bins=15,
        color=['coral', 'royalblue', 'purple'],
        label=['data 1', 'data 2', 'data 3'])

ax.tick_params(labelsize=15)
ax.legend(fontsize=20)
```

#### returns of 2D data

```python
n_class, n_data = 2, 300

# data's shape: (300, 2)
data = np.random.normal(0, 1, (n_data, n_class))

fig, ax = plt.subplots(figsize=(14, 10))
freqs, bin_edges, rects = ax.hist(data)
print(freqs.shape)
print(bin_edges.shape)
print(rects)
# rects[0], rects[1]로 접근 가능

ax.tick_params(labelsize=15)
```

```python
n_class, n_data = 2, 300

# data's shape: (300, 2)
data = np.random.normal(0, 1, (n_data, n_class))

fig, ax = plt.subplots(figsize=(14, 10))
freqs, bin_edges, rects = ax.hist(data)

bin_center = (bin_edges[:-1] + bin_edges[1:]) / 2
ax.set_xticks(bin_center)
ax.set_xticklabels(np.around(bin_center, 2),
                   rotation=30)

# annotation
for rect in rects[0]:
    x = rect.get_x()
    width = rect.get_width()
    height = rect.get_height()
    
    ax.text(x + .5*width, height+3,
            str(height),
            rotation=90,
            fontsize=15,
            ha='center',
            color='b')
for rect in rects[1]:
    x = rect.get_x()
    width = rect.get_width()
    height = rect.get_height()
    
    ax.text(x + .5*width, height+3,
            str(height),
            rotation=90,
            fontsize=15,
            ha='center',
            color='r')

ylim = ax.get_ylim()
ax.set_ylim([ylim[0], ylim[1]+10])

ax.tick_params(labelsize=15)
```

```python
# 다른 annotation 방법

n_class, n_data = 2, 300

# data's shape: (300, 2)
data = np.random.normal(0, 1, (n_data, n_class))

fig, ax = plt.subplots(figsize=(14, 10))
freqs, bin_edges, rects = ax.hist(data)

bin_center = (bin_edges[:-1] + bin_edges[1:]) / 2
ax.set_xticks(bin_center)
ax.set_xticklabels(np.around(bin_center, 2),
                   rotation=30)

# annotation
colors = ['b', 'r']
for class_idx in range(n_class):
    for rect in rects[class_idx]:
        x = rect.get_x()
        width = rect.get_width()
        height = rect.get_height()
        
        ax.text(x + .5*width, height + 3,
                str(height),
                rotation=90,
                fontsize=15,
                ha='center',
                color=colors[class_idx])

ylim = ax.get_ylim()
ax.set_ylim([ylim[0], ylim[1]+10])

ax.tick_params(labelsize=15)
```
