## 1) ax.hist

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
ax.hist(data)

ax.tick_params(labelsize=20,
               length=10,
               width=3)
```

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
ax.hist(data, facecolor='skyblue')

ax.tick_params(labelsize=20,
               length=10,
               width=3)
```

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
ax.hist(data,
        facecolor='skyblue',
        edgecolor='darkblue',
        linewidth=3)

ax.tick_params(labelsize=20,
               length=10,
               width=3)
```

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
ax.hist(data,
        facecolor='skyblue',
        edgecolor='darkblue',
        linewidth=3,
        hatch='//')

ax.tick_params(labelsize=20,
               length=10,
               width=3)
```

## 2) ax.hist 리턴값
- freqs
- bin_edges
- rects

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
freqs, bin_edges, rects = ax.hist(data)
print('freqs(각 빈도 횟수): ', freqs)
print('bin_edges: ', bin_edges)
# rects에는 bar object 10개가 들어있다. (same as before)

for rect in rects:
    x = rect.get_x()
    width = rect.get_width()
    height = rect.get_height()
    
    ax.text(x + .5*width, height + 2,
            str(height),
            fontsize=10,
            ha='center')

ax.tick_params(labelsize=20,
               length=10,
               width=3)
```

### bin_edges

```python
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
freqs, bin_edges, rects = ax.hist(data)

ax.set_xticks(bin_edges)
ax.set_xticklabels(np.around(bin_edges, 2),
                   rotation=30)

ax.tick_params(labelsize=15,
               length=10,
               width=3)
```

```python
# bin_edges가 아닌 bin_centers으로 설정(꿀팁🍯)
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
freqs, bin_edges, rects = ax.hist(data)

bin_centers = (bin_edges[:-1] + bin_edges[1:])/2

ax.set_xticks(bin_centers)
ax.set_xticklabels(np.around(bin_centers, 2),
                   rotation=30)

ax.tick_params(labelsize=15,
               length=10,
               width=3)
```

```python
# 히스토그램에 대한 추가적인 정보 제공
np.random.seed(0)

n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 7))
freqs, bin_edges, rects = ax.hist(data)

n_bin = len(freqs)
bin_edges_around = np.around(bin_edges, 2)

m_val, M_val = bin_edges_around[0], bin_edges_around[-1]
bin_interval = np.around(bin_edges[1] - bin_edges[0], 2)

text_loc = ax.get_xlim()[0], ax.get_ylim()[1]
text_template = 'Number of Bins: {}\n' \
    'Min, Max Values: {} / {} \n' \
    'Bin Interval: {}'

ax.text(text_loc[0] + .1, text_loc[1] - 5,
        text_template.format(n_bin,
                             m_val, M_val,
                            bin_interval),
        ha='left', va='top',
        fontsize=10)

ax.tick_params(labelsize=15,
               length=10,
               width=3)
```
