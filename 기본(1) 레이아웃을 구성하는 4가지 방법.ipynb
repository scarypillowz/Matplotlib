# 4가지 레이아웃 구성법

```python
import matplotlib.pyplot as plt
```

## ✅ plt.figure
- 그래프/이미지를 그리기 위한 전체 영역을 확보한다. 일종의 window.
- figure 객체를 인스턴스화해서 사용하는 것을 추천한다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='gray')
```

## ✅ fig.add_subplot
- 하나의 그래프/이미지를 위한 영역(ax)을 생성한다. (그래프/이미지당 하나의 ax를 갖는 것)
- 그리드 시스템을 바라보는 관점을 다르게 하면 레이아웃에 변화를 줄 수 있다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='moccasin')

ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='moccasin')

# 2행 2열 그리드 시스템의 첫 번째 ax
ax1 = fig.add_subplot(2, 2, 1)

# 2행 2열 그리드 시스템의 두 번째 ax
ax2 = fig.add_subplot(2, 2, 2)

# 2행 1열 그리드 시스템의 두 번째 ax
ax3 = fig.add_subplot(2, 1, 2)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='moccasin')

ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(133)
ax4 = fig.add_subplot(234)
ax5 = fig.add_subplot(235)
```

## ✅ plt.subplots
- plt.subplots = plt.figure + fig.add_subplot
- 규칙적인 레이아웃은 구현할 수 있지만 위와 같은 불규칙적 레이아웃은 구현이 불가능하다.
- axes.flat (다차원 배열을 1차원으로 평탄화한다) ⭐️⭐️⭐️⭐️⭐️
- 자주 사용하는 레이아웃 구성법이다.

```python
# fig, (ax1, ax2) = plt.subplots(2, 1)
fig, axes = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(7, 7),
    facecolor='lightcoral'
)
```

```python
# axes는 넘파이 배열이다.
print(type(axes))
print('axes[0]: ', axes[0])
print('axes[1]: ', axes[1])
```

```python
fig, axes = plt.subplots(
    2, 2,
    figsize=(7, 7),
    facecolor='lightcoral'
)
print('axes shape: ', axes.shape)
print(axes)

# 다차원 배열을 1차원으로 평탄화
for ax in axes.flat:
    print(ax)
    
for idx, ax in enumerate(axes.flat):
    print(idx, ax)
```

### flat을 사용하는 이유

```python
title_lists = ['Training Image', '2 Scales', '4 Scales',
              '5 Scales', '6 Scales', '8 Scales']
fig, axes = plt.subplots(2, 3, figsize=(15, 10), facecolor='lightcoral')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_lists[ax_idx], fontsize=15)
```

## ✅ plt.subplot2grid
- fig.add_subplot, plt.subplots 메서드로는 구현하기 복잡한 레이아웃을 구성할 수 있다.
- figure 객체를 인수로 전달해야 한다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='yellowgreen')

# 2행 1열의 레이아웃에서 (0, 0) 위치에 그린다.
ax1 = plt.subplot2grid((2, 1), (0, 0), fig=fig)

# 2행 1열의 레이아웃에서 (1, 0) 위치에 그린다.
ax2 = plt.subplot2grid((2, 1), (1, 0), fig=fig)
```

```python
# 2차원 레이아웃
fig = plt.figure(figsize=(7, 7), facecolor='yellowgreen')

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2, fig=fig)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=3, fig=fig)
ax3 = plt.subplot2grid((3, 3), (2, 0), fig=fig)
ax4 = plt.subplot2grid((3, 3), (2, 1), fig=fig)
ax5 = plt.subplot2grid((3, 3), (2, 2), fig=fig)
```

```python
# fig.add_subplot()과 plt.subplot2grid() 메서드를 사용해서 동일한 결과를 얻을 수 있다.
fig = plt.figure(figsize=(7, 7), facecolor='moccasin')
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 2, 3)
ax3 = fig.add_subplot(2, 2, 4)

fig2 = plt.figure(figsize=(7, 7), facecolor='yellowgreen')
ax4 = plt.subplot2grid((2, 2), (0, 0), colspan=2, fig=fig2)
ax5 = plt.subplot2grid((2, 2), (1, 0), fig=fig2)
ax6 = plt.subplot2grid((2, 2), (1, 1), fig=fig2)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='yellowgreen')

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2, fig=fig)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2, fig=fig)
ax3 = plt.subplot2grid((3, 3), (2, 0), colspan=2, fig=fig)
ax4 = plt.subplot2grid((3, 3), (0, 2), rowspan=3, fig=fig)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='yellowgreen')

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3, fig=fig)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2, fig=fig)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2, fig=fig)
ax4 = plt.subplot2grid((3, 3), (2, 0), fig=fig)
ax5 = plt.subplot2grid((3, 3), (2, 1), fig=fig)
```

## ✅ fig.add_axes
- fig.add_axes([left, bottom, width, height])
    - left: 왼쪽으로부터 떨어진 거리
    - bottom: 밑으로부터 떨어진 거리
    - width: (1에 대한 비율값) 가로 길이
    - height: (1에 대한 비율값) 세로 길이
- 중첩 그래프/이미지를 그릴 수 있다.
- auto-scaled 레이아웃을 구현할 수 있다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='skyblue')

rect1 = [0.1, 0.1, 0.5, 0.5]
rect2 = [0.7, 0.1, 0.2, 0.5]
rect3 = [0.1, 0.7, 0.8, 0.2]

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```

```python
# 개선된 코드
fig = plt.figure(figsize=(7, 7), facecolor='skyblue')

left, bottom = 0.1, 0.1
spacing = 0.05
width1, height1 = 0.5, 0.5
width2 = 1 - (2*left + spacing + width1)
height2 = 1 - (2*bottom + spacing + height1)

rect1 = [left, bottom , width1, height1]
rect2 = [left, bottom + height1 + spacing, 1 - left*2, height2]
rect3 = [left + width1 + spacing, bottom, width2, height1]

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```

```python
# 스케일만 조정하면 ax의 크기를 조절할 수 있다.
fig = plt.figure(figsize=(7, 7), facecolor='skyblue')

left, bottom = 0.1, 0.1
spacing = 0.05
width1, height1 = 0.3, 0.3
width2 = 1 - (2*left + spacing + width1)
height2 = 1 - (2*bottom + spacing + height1)

rect1 = [left, bottom , width1, height1]
rect2 = [left, bottom + height1 + spacing, 1 - left*2, height2]
rect3 = [left + width1 + spacing, bottom, width2, height1]

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```
