```python
np.random.seed(0)
y_data = np.random.normal(0, 1, (300, ))

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(y_data)

fig.tight_layout(pad=3)

x_ticks = np.arange(301, step=50)
ax.set_xticks(x_ticks)
ax.tick_params(labelsize=15)

ax.grid()

for spine_loc, spine in ax.spines.items():
    spine.set_linewidth(3)
    if spine_loc in ['right', 'top']:
        spine.set_visible(False)
    if spine_loc in ['left', 'bottom']:
        spine.set_position(('data', 0))
```

## 6. line styles & markers

### ‚úÖ markers

```python
PI = np.pi
t = np.linspace(-4*PI, 4*PI, 30)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(t, sin, label='sin(t)')
ax.plot(t, sin+1, label='sin(t)+1', marker='o')
ax.plot(t, sin+2, label='sin(t)+2', marker='D')
ax.plot(t, sin+3, label='sin(t)+3', marker='s')

ax.legend(loc='center left', bbox_to_anchor=(1, .5), fontsize=15)

fig.tight_layout()
```

### ‚úÖ Q5

```python
# Îç∞Ïù¥ÌÑ∞
name_list = ['PL', 'NL', 'NL<SeINL', 'NL->SeINL->SeIPL']
x_data = np.linspace(.5, 1, 30)
y_data = np.empty(shape=(0, len(x_data)))
gamma_list = [6, 7, 8, 12]
for i in range(len(gamma_list)):
    y = (-.5)*np.power(x_data, gamma_list[i]) + 1
    y = y.reshape(1, len(x_data))
    y_data = np.vstack([y_data, y])
    
# plotting
fig, ax = plt.subplots(figsize=(8, 8))
for i in range(len(name_list)):
    ax.plot(x_data, y_data[i], label=name_list[i],
            marker='o', markersize=8)
    
ax.set_xlim([.5, 1.0])
ax.set_ylim([.5, 1.0])

label_dict = {'size': 15}
ax.set_xlabel('recall', fontdict=label_dict)
ax.set_ylabel('precision', fontdict=label_dict)

ax.grid(linewidth=1.5, alpha=.5)
    
# spine
for spine_loc, spine in ax.spines.items():
    spine.set_linewidth(2)
    spine.set_alpha(.5)
    
# legend
ax.legend(loc='lower left', fontsize=12)

# tick, ticklabel
tick_dict = {'size': 15}
ticks = [round(0.1*i, 1) for i in range(5, 11)]

ax.set_xticks(ticks)
ax.set_xticklabels(ticks, fontdict=tick_dict)
ax.set_yticks(ticks)
ax.set_yticklabels(ticks, fontdict=tick_dict);
```

```python
# Îç∞Ïù¥ÌÑ∞
data_loc = [46, 58, 65, 73, 80, 92, 96, 98]

supervised_data = [60, 65, 63, 80, 68, 70, 78, 85]
supervised_top_data = [98, 110, 90, 150, 125, 120, 130, 140]
supervised_bottom_data = [40, 30, 45, 30, 40, 40, 40, 40]

distilled_data = [25, 26, 15, 30, 24, 44, 49, 51]
distilled_top_data = [48, 55, 41, 48, 44, 55, 52, 86]
distilled_bottom_data = [5, 0, 7.5, 15, 10, 0, 20, 17]

# plotting
fig, ax = plt.subplots(figsize=(13, 10))

ax.fill_between(data_loc,
                y1=supervised_top_data,
                y2=supervised_bottom_data,
                color='dodgerblue',
                alpha=.4)
ax.fill_between(data_loc,
                y1=distilled_top_data,
                y2=distilled_bottom_data,
                color='peachpuff')

ax.plot(data_loc, supervised_data,
        marker='x',
        color='dodgerblue',
        linewidth=3,
        markersize=15,
        label='Supervised Student')
ax.plot(data_loc, distilled_data,
        marker='o',
        markersize=15,
        markerfacecolor='peachpuff',
        markeredgewidth=3,
        color='darkorange',
        label='Distilled Student')

ax.axhline(y=25, color='black',
           linestyle='--', linewidth=3,
           # labelÏùÄ legendÏùò Î™©Î°ùÏùÑ ÏùòÎØ∏ÌïúÎã§.
           label='Supervised Teacher', alpha=.5)

# axis limit
ax.set_xlim([45, 100])
ax.set_ylim([0,  200])

# spine
for spine_loc, spine in ax.spines.items():
    if spine_loc in ['right', 'top']:
        spine.set_visible(False)
    if spine_loc in ['left', 'bottom']:
        spine.set_linewidth(1.8)
        
# tick, ticklabels
x_ticks = np.arange(50, 91, 10)
y_ticks = np.arange(0, 201, 50)

ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)

ax.tick_params(direction='in',
               length=5,
               width=1.8,
               labelsize=20)

# legend
ax.legend(loc='upper left',
          fontsize=17)

ax.set_xlabel('Distillation Rate (%)', fontsize=25)
ax.set_ylabel('RMS Absolute Pose Errors (m)', fontsize=25);
```
