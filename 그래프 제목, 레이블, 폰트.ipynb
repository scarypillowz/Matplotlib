```python
import matplotlib.pyplot as plt
```

# 그래프 제목, 레이블, 폰트

## 1) 제목
- 2가지 제목
    - figure 제목(fig.suptitle)
    - ax 제목(ax.set_title)

### 1-1) figure 제목
- 자주 사용하지는 않는다.
- subplots_adjust() 메서드를 사용할 것.

```python
fig, ax = plt.subplots(figsize=(5, 5))
fig.suptitle('Title of a Figure', fontsize=20, fontfamily='monospace')
```

```python
ax_titles = ['title' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(7, 7))
fig.suptitle('Title of a Figure', fontsize=40, fontfamily='monospace')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(ax_titles[ax_idx], fontsize=30, fontfamily='serif')
    
fig.tight_layout()
```

```python
ax_titles = ['title' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(7, 7), facecolor='linen')
fig.suptitle('Title of a Figure', fontsize=40, fontfamily='monospace')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(ax_titles[ax_idx], fontsize=30, fontfamily='serif')
    
fig.subplots_adjust(bottom=0.05, top=0.8, hspace=0.3)
```

### 1-2) ax 제목

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_title('Title of a ax')
```

## 2) 레이블
- 2가지 레이블
    - x축 레이블(ax.set_xlabel)
    - y축 레이블(ax.set_ylabel)
- fontsize, color, alpha 등

```python
fig, ax = plt.subplots(figsize=(7, 7))

ax.set_xlabel('X label', fontsize=20, color='darkblue', alpha=.7)
ax.set_ylabel('Y label', fontsize=20, color='darkblue', alpha=.7)
```

```python
# 타이틀과 레이블
fig, ax = plt.subplots(figsize=(7, 7))

fig.suptitle('Title of a Figure', fontsize=30, color='darkblue', alpha=.9)
ax.set_xlabel('X label', fontsize=20, color='darkblue', alpha=.7)
ax.set_ylabel('Y label', fontsize=20, color='darkblue', alpha=.7)
```

## 3) 연습문제

```python
fig, ax = plt.subplots(figsize=(7, 7))

ax_impulse = fig.add_axes([.2, .55, .25, .25])
ax_prob = fig.add_axes([.6, .55, .25, .25])

ax_impulse.get_xaxis().set_visible(False)
ax_impulse.get_yaxis().set_visible(False)
ax_prob.get_xaxis().set_visible(False)
ax_prob.get_yaxis().set_visible(False)

ax.set_title('Gaussian colored noise', fontsize=20)
ax.set_xlabel('time (s)', fontsize=15)
ax.set_ylabel('current (nA)', fontsize=15)

ax_impulse.set_title('Impulse response', fontsize=15)
ax_prob.set_title('Probability', fontsize=15)
```
