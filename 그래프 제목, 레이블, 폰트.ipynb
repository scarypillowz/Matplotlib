```python
import matplotlib.pyplot as plt
```

# 그래프 제목, 레이블, 폰트

## 1) 제목
- 2가지 제목
    - figure 제목(fig.suptitle)
    - ax 제목(ax.set_title)

### 1-1) figure 제목
- 자주 사용하지는 않는다.
- subplots_adjust() 메서드를 사용할 것.

```python
fig, ax = plt.subplots(figsize=(5, 5), facecolor='linen')
fig.suptitle('Title of a Figure', fontsize=20, fontfamily='monospace')
```

```python
ax_titles = ['title' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(7, 7), facecolor='linen')
fig.suptitle('Title of a Figure', fontsize=40, fontfamily='monospace')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(ax_titles[ax_idx], fontsize=30, fontfamily='serif')
    
fig.tight_layout()
```

```python
ax_titles = ['title' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(7, 7), facecolor='linen')
fig.suptitle('Title of a Figure', fontsize=40, fontfamily='monospace')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(ax_titles[ax_idx], fontsize=30, fontfamily='serif')
    
fig.subplots_adjust(bottom=0.05, top=0.8, hspace=0.3)
```

### 1-2) ax 제목

```python
fig, ax = plt.subplots(figsize=(5, 5), facecolor='linen')
ax.set_title('Title of a ax')
```

## 2) 레이블
- 2가지 레이블
    - x축 레이블(ax.set_xlabel)
    - y축 레이블(ax.set_ylabel)
- fontsize, color, alpha 등

```python
fig, ax = plt.subplots(figsize=(7, 7), facecolor='linen')

ax.set_xlabel('X label', fontsize=20, color='darkblue', alpha=.7)
ax.set_ylabel('Y label', fontsize=20, color='darkblue', alpha=.7)
```

```python
# 타이틀과 레이블
fig, ax = plt.subplots(figsize=(7, 7), facecolor='linen')

fig.suptitle('Title of a Figure', fontsize=30, color='darkblue', alpha=.9)
ax.set_xlabel('X label', fontsize=20, color='darkblue', alpha=.7)
ax.set_ylabel('Y label', fontsize=20, color='darkblue', alpha=.7)
```

## 3) 연습문제

```python
fig, ax = plt.subplots(figsize=(7, 7), facecolor='linen')

ax_impulse = fig.add_axes([.2, .55, .25, .25])
ax_prob = fig.add_axes([.6, .55, .25, .25])

ax_impulse.get_xaxis().set_visible(False)
ax_impulse.get_yaxis().set_visible(False)
ax_prob.get_xaxis().set_visible(False)
ax_prob.get_yaxis().set_visible(False)

ax.set_title('Gaussian colored noise', fontsize=20)
ax.set_xlabel('time (s)', fontsize=15)
ax.set_ylabel('current (nA)', fontsize=15)

ax_impulse.set_title('Impulse response', fontsize=15)
ax_prob.set_title('Probability', fontsize=15)
```

## 4) text alignment
- horizontal alignment
- vertical alignment
- 기준점 설정

### 4-1) horizontal alignment

```python
fig, ax = plt.subplots(figsize=(7, 7))

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

ax.grid()
ax.tick_params(axis='both', labelsize=15)

# ax.text(x=0, y=0, s='Hello', va='center', ha='left', fontsize=30)
# ax.text(x=0, y=0, s='Hello', va='center', ha='center', fontsize=30)
ax.text(x=0, y=0, s='Hello', va='center', ha='right', fontsize=30)
```

### 4-2) vertical alignment

```python
fig, ax = plt.subplots(figsize=(7, 7))

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

ax.grid()
ax.tick_params(axis='both', labelsize=15)

# ax.text(x=0, y=0, s='Hello', va='top', ha='center', fontsize=30)
ax.text(x=0, y=0, s='Hello', va='bottom', ha='center', fontsize=30)
```

## 5) title alignment
- ax.set_title(..., y)
    - y는 음수 가능

```python
ax_titles = ['(a) CE - clean', '(b) CE - noisy', '(c) LSR - noisy', '(d) SL - noisy']

title_bottom = .9
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

fig.suptitle('Symmetric Cross Entropy', fontsize=30, y=title_bottom + .03, va='bottom')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(ax_titles[ax_idx], fontsize=20, y=-0.25)
    ax.set_xlabel('Number of Epochs', fontsize=15)
    ax.set_ylabel('Class-wise Test Accuracy', fontsize=15)

fig.subplots_adjust(top=title_bottom, bottom=.1, hspace=.3, wspace=.3)
```
