## 1) ax.tick_params()
- tick Í¥ÄÎ†® Í∏∞Î≥∏ ÏÑ§Ï†ï
- ax.set_xticks, ax.set_yticks, ax.set_xticklabels, ax.set_yticklabels Ïù¥ÌõÑÏóê ÏÑ§Ï†ï
- Îß§Í∞úÎ≥ÄÏàò
    - axis: 'x', 'y', 'both' (ÎîîÌè¥Ìä∏ 'both')
    - which: 'major', 'minor', 'both' (ÎîîÌè¥Ìä∏ 'major')
    - labelsize: Ìã± Ìè∞Ìä∏ ÌÅ¨Í∏∞
    - length: Ìã± Í∏∏Ïù¥(figure ÏÇ¨Ïù¥Ï¶àÍ∞Ä ÌÅ¨Î©¥ Í∞íÏù¥ ÌÅ¥ ÌïÑÏöîÍ∞Ä ÏûàÎã§)
    - width: Ìã± ÎÑàÎπÑ(figure ÏÇ¨Ïù¥Ï¶àÍ∞Ä ÌÅ¨Î©¥ Í∞íÏù¥ ÌÅ¥ ÌïÑÏöîÍ∞Ä ÏûàÎã§)
    - bottom, top, left, right: Ìã± ÌëúÏãú Ïó¨Î∂Ä(Ïù¥ÎØ∏ÏßÄ Îã§Î£∞ Îïå ÏÇ¨Ïö©ÌïúÎã§)
    - labelbottom, labeltop, labelleft, labelright: Ìã± Î†àÏù¥Î∏î ÌëúÏãú Ïó¨Î∂Ä(Ïù¥ÎØ∏ÏßÄ Îã§Î£∞ Îïå ÏÇ¨Ïö©ÌïúÎã§)
    - labelrotation
    - direction: Ìã±Ïùò ÏúÑÏπò(ax ÎÇ¥Î∂Ä ÌòπÏùÄ Ïô∏Î∂Ä)
    - colors: Ìã±Í≥º Ìã± Î†àÏù¥Î∏î ÏÉâ

## 2) ax.set_xticks(), ax.set_yticks()
- ax.set_xlim() Î©îÏÑúÎìúÎäî xticksÎ•º Í∑†ÏùºÌïòÍ≤å ÏûêÎèô ÏÉùÏÑ±ÌïúÎã§.
- ÌïòÏßÄÎßå ax.set_xticks() Î©îÏÑúÎìúÎäî Ï†ÑÎã¨Ìïú Î∞∞Ïó¥ ÏïàÏùò Í∞íÎì§Î°ú xticksÎ•º ÏÉùÏÑ±ÌïúÎã§.
- Îß§Í∞úÎ≥ÄÏàò
    - minor: ÎîîÌè¥Ìä∏Îäî False. Ï¶â Í∏∞Î≥∏Ï†ÅÏúºÎ°úÎäî major tickÏùÑ Í±¥ÎìúÎ¶∞Îã§.

```python
fig, ax = plt.subplots(figsize=(5, 5), facecolor='lightgreen')

xticks = [i for i in range(0, 101, 20)]

ax.set_xticks(xticks)
ax.tick_params(axis='x', labelsize=10, length=10, width=3, labelrotation=30)
```

```python
fig, ax = plt.subplots(figsize=(14, 7), facecolor='lightgreen')

major_xticks = [i for i in range(0, 101, 20)]
minor_xticks = [i for i in range(0, 101, 5)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks, minor=True)
ax.tick_params(axis='x', labelsize=15, length=10, width=3, rotation=30)
ax.tick_params(axis='x', which='minor', labelsize=8, length=5, width=1.5)
```

```python
fig, ax = plt.subplots(figsize=(14, 7), facecolor='lightgreen')

major_xticks = [i for i in range(0, 101, 20)]
minor_xticks = [i for i in range(0, 101, 5)]
major_yticks = [i for i in range(0, 11, 2)]
minor_yticks = [i for i in range(0, 11, 1)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks, minor=True)
ax.tick_params(axis='x', labelsize=15, length=10, width=3, rotation=30)
ax.tick_params(axis='x', which='minor', length=5, width=1.5)

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)
ax.tick_params(axis='y', labelsize=15, length=10, width=3)
ax.tick_params(axis='y', which='minor', length=5, width=1.5)
```

## 3) ax.set_xticklabels(), ax.set_yticklabels()
- ax.set_xticks(), ax.set_yticks() Î©îÏÑúÎìú Ïù¥ÌõÑÏóê ÏûëÏÑ±Ìï¥Ïïº ÌïúÎã§. ÏàúÏÑúÏóê Ïú†Ïùò!
    - ‚≠êÔ∏è set_ticks() > set_ticklabels() > tick_params()
- Îß§Í∞úÎ≥ÄÏàò
    - minor: ÎîîÌè¥Ìä∏Îäî False.

```python
xticks = [i for i in range(10)]
xtick_labels = ['Class ' + str(i) for i in xticks]

yticks = [i for i in range(0, 101, 20)]
ytick_labels = [str(i) + '%' for i in yticks]

fig, ax = plt.subplots(figsize=(5, 5), facecolor='lightyellow')

ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels)

ax.set_yticks(yticks)
ax.set_yticklabels(ytick_labels)

ax.tick_params(axis='x', labelsize=10, labelrotation=60)
ax.tick_params(axis='y', labelsize=10)
fig.subplots_adjust(left=.15, bottom=.2)
```

```python
xticks = [i for i in range(7)]
xtick_labels = [' ',
               '<vehicle, move along, street>',
               '<person, move along, sidewalk>',
               '<pet, move along, sidewalk>',
               '<person, stay, lawn>',
               '<person, move away (home), driveway>',
               ' ']
yticks = [i/10 for i in range(11)]

fig, ax = plt.subplots(figsize=(20, 10), facecolor='lightyellow')

ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels, ha='right')
ax.set_yticks(yticks)

ax.tick_params(axis='x', rotation=20, labelsize=15, colors='gray', width=2, length=8)
ax.tick_params(axis='y', labelsize=15, colors='gray')

fig.subplots_adjust(bottom=.2, top=.95, left=.1, right=.9)
```

```python
xticks = [i for i in range(8)]
yticks = np.arange(0.4, 0.54, 0.02)

xtick_labels = [' ',
                'FC-Agg 1 layer',
                'FC-Agg 2 layer',
                'Topo-Agg 0-hop',
                'Topo-Agg 0-hop test only',
                'Topo-Agg 1-hop',
                'Topo-Agg 2-hop',
                ' ']

fig, ax = plt.subplots(figsize=(20, 10), facecolor='lightyellow')

# xÏ∂ï(ÏàúÏÑúÏóê Ïú†Ïùò)
ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels, ha='right')
# yÏ∂ï
ax.set_yticks(yticks)
ax.set_ylim([0.4, 0.52])

ax.tick_params(axis='x', rotation=30, labelsize=15, colors='gray')
ax.tick_params(axis='y', labelsize=15, colors='gray')
```

```python
fig, ax = plt.subplots(figsize=(10, 10), facecolor='lightyellow')
ax.set_xscale('log')
ax.set_yscale('log')

major_xticks = [10**i for i in range(5)]
major_yticks = [1E-10, 1E-5, 1E0]
minor_yticks = [10**i for i in range(-10, 4)]
ax.set_xticks(major_xticks)
ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)

ax.tick_params(which='major', direction='in', length=8, labelsize=20)
ax.tick_params(which='minor', direction='in', length=5, labelsize=0)
```

- Í∏∞Î≥∏Ï†ÅÏúºÎ°ú major ticksÏóê ÎåÄÌï¥ÏÑú gridÎ•º Í∑∏Î¶∞Îã§.
- Ï†ïÎ≥¥ Ï†ÑÎã¨Î†•ÏùÑ ÎÜíÏùº Ïàò ÏûàÎã§.

## 2) ticks & grid

```python
fig, ax = plt.subplots(figsize=(7, 7), facecolor='lightblue')

major_yticks = [i for i in range(0, 101, 20)]
minor_yticks = [i for i in range(0, 101, 5)]

ax.set_yticks(major_xticks)
ax.set_yticks(minor_xticks, minor=True)

ax.tick_params(labelsize=15)

# ÏúÑÏóêÏÑú ÏÑ§Ï†ïÌïú ticksÏóê ÎåÄÌï¥ gridÎ•º Í∑∏Î†§Ï§ÄÎã§.
ax.grid(axis='y', which='major', linewidth=1.5)
ax.grid(axis='y', which='minor', linestyle='-.')
```

```python
fig, ax = plt.subplots(figsize=(7, 7), facecolor='lightblue')

major_yticks = [i for i in range(0, 101, 20)]
minor_yticks = [i for i in range(0, 101, 5)]

ytick_labels = [str(i) + '%' for i in major_yticks]

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)
ax.set_yticklabels(ytick_labels)

ax.tick_params(labelsize=12)

ax.grid(axis='y', which='major', linewidth=1.5)
ax.grid(axis='y', which='minor', linestyle=':')
```
