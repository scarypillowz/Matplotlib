```python
import matplotlib.pyplot as plt
import numpy as np
```

# 그래프 spines
- 4개의 면을 의미한다.
- tick, ticklabel과는 별개다.
- ax.spines.items() 메서드와 반복문을 사용해서 접근한다.
- Spine 객체의 메서드
    - set_visible()
    - set_position()

```python
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    print(spine_loc, spine, sep=' - ')
    
# 조건 연산
for spine_loc, spine in ax.spines.items():
    if spine_loc in ['left', 'bottom']:
        pass
    
    if spine_loc in ['right', 'top']:
        pass
```

## 1) spine.set_visible()

```python
# 반복문을 사용한다.
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['top', 'right']:
        spine.set_visible(False)
```

```python
# tick, ticklabel과는 별개
fig, ax = plt.subplots(figsize=(5, 5))
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.tick_params(left=False, labelbottom=False)
```

```python
# spines, ticks 함께 사용하기
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['left', 'bottom']:
        spine.set_visible(False)
        
ax.tick_params(left=False, labelleft=False,
               right=True, labelright=True,
               bottom=False, labelbottom=False,
               top=True, labeltop=True)
```

```python
# 이미지 그리기
np.random.seed(0)

noise_img = np.random.normal(0, 1, (100, 100))

fig, ax = plt.subplots(figsize=(5, 5))
ax.imshow(noise_img)

for spine_loc, spine in ax.spines.items():
    spine.set_visible(False)
    
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

## 2) spine.set_position()
- 매개변수
    - axes
    - data

```python
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['bottom', 'right', 'top']:
        spine.set_visible(False)

ax.tick_params(bottom=False, labelbottom=False)

# 💀
ax.spines['left'].set_position('center')
```

```python
# 💀 (set_position 메서드의 값 'center'는 말 그대로 spine을 중앙으로 이동시킨다는 것을 의미한다.)
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['right', 'top']:
        spine.set_visible(False)
        
    if spine_loc in ['left', 'bottom']:
        spine.set_position('center')
```

```python
# 'center'의 의미는 ('axes', 0.5)와 동일하다.
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['bottom', 'right', 'top']:
        spine.set_visible(False)
        
ax.tick_params(bottom=False, labelbottom=False)

# ax.spines['left'].set_position('center')
ax.spines['left'].set_position(('axes', 0.5))
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['bottom', 'right', 'top']:
        spine.set_visible(False)
        
ax.tick_params(bottom=False, labelbottom=False)

ax.spines['left'].set_position(('axes', 0.8))
```

```python
# 😇 desirable way
x = np.random.normal(2, 3, (100, ))
y = np.random.normal(2, 3, (100, ))

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(x, y)

ax.tick_params(labelsize=15)
fig.tight_layout()

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['right', 'top']:
        spine.set_visible(False)
        
    if spine_loc in ['left', 'bottom']:
        spine.set_position(('data', 0))
```

```python
x = np.random.normal(5, 3, (500, ))
y = np.random.normal(3, 3, (500, ))

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x, y)

ax.tick_params(labelsize=15, colors='red')
fig.tight_layout()

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['right', 'top']:
        spine.set_visible(False)
    
    if spine_loc == 'left':
        spine.set_position(('data', 5))
        spine.set_color('red')
        
    if spine_loc == 'bottom':
        spine.set_position(('data', 3))
        spine.set_color('red')
```
