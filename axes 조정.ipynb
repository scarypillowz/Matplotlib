```python
import matplotlib.pyplot as plt
```

# axes 조정하기

## 1) fig.tight_layout()
- fig.tight_layout(pad)
    - pad: 간격 설정
    - ⭐️ ax 사이 간격과 fig 상하좌우 간격을 한꺼번에 설정
- fig.tight_layout() 메서드는 title + xlabel + ylabel 을 포함해서 레이아웃을 구성한다.
- 레이블명과 타이틀이 겹치지 않기 때문에 깔끔한 figure를 얻을 수 있다.
- 메서드는 figure 객체에 적용한다.

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.set_title('This is Title', fontsize=20)
ax.set_xlabel('X Label', fontsize=15)
ax.set_ylabel('Y Label', fontsize=15)
```

```python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['X label' + str(i) for i in range(4)]
ylabel_list = ['Y label' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(7, 7))
for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx], fontsize=30)
    ax.set_xlabel(xlabel_list[ax_idx], fontsize=20)
    ax.set_ylabel(ylabel_list[ax_idx], fontsize=20)
    
fig.tight_layout(pad=6)  # ⭐️
```

## 2) axis.set_visible()
- axis 눈금을 없애기 때문에 이미지를 그릴 때 유용하다.

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```

## 3) fig.subplots_adjust()
- 좌우/상하 간격을 설정할 수 있다.
    - ⭐️ ax 사이 간격과 fig 상하좌우 간격을 각각 설정
- 총 6개의 매개변수를 전달할 수 있다.
    - positional (fig 상하좌우 간격)
        - left, right, bottom, top
        - 전체 axes들이 차지하는 위치를 설정할 수 있다. (As a whole)
    - spacing (ax 사이 간격)
        - wspace, hspace

### 3-1) positional arguments

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom=0.01, top=0.99)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(left=0.01, right=0.99)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom=0.01, top=0.99, left=0.01, right=0.99)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom=0.05, top=0.9, left=0.05, right=0.8)
```

### 3-2) spacing arguments

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(hspace=0.05)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(wspace=0.05)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(10, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

fig.subplots_adjust(hspace=0.2, wspace=0.05)
```

### 3-3) 연습문제

```python
import numpy as np
fig = plt.figure(figsize=(12, 13))

# axes 그룹1
axes_g1 = np.empty(shape=(0, ))
main_g1 = plt.subplot2grid((5, 4), (0, 0), 2, 2, fig=fig)
main_g1.get_xaxis().set_visible(False)
main_g1.get_yaxis().set_visible(False)

axes_g1 = np.append(axes_g1, main_g1)
for r_idx in range(2, 5):
    for c_idx in range(2):
        ax = plt.subplot2grid((5, 4), (r_idx, c_idx), fig=fig)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        axes_g1 = np.append(axes_g1, ax)
axes_g1 = axes_g1.reshape(1, -1)        

# axes 그룹2
axes_g2 = np.empty(shape=(0, ))
main_g2 = plt.subplot2grid((5, 4), (0, 2), 2, 2, fig=fig)
main_g2.get_xaxis().set_visible(False)
main_g2.get_yaxis().set_visible(False)

axes_g2 = np.append(axes_g2, main_g2)
for r_idx in range(2, 5):
    for c_idx in range(2, 4):
        ax = plt.subplot2grid((5, 4), (r_idx, c_idx), fig=fig)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        axes_g2 = np.append(axes_g2, ax)
axes_g2 = axes_g2.reshape(1, -1)

fig.subplots_adjust(bottom=0.03, top=0.97, left=0.03, right=0.97, hspace=0, wspace=0)
```
