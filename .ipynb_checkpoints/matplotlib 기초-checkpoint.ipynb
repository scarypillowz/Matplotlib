```python
import matplotlib.pyplot as plt
```

# matplotlib 기초

## 1. plt.figure()
- 그래프를 그리기 위한 전체 영역을 확보한다.
- fig 객체를 인스턴스화 한 뒤 이를 사용한다

```python
fig1 = plt.figure(figsize=(7, 7), facecolor='gray')
```

## 2. fig.add_subplot()
- 그래프만의 영역인 ax를 생성한다.
- 생성한 ax에 그래프를 그릴 수 있다.

```python
ax1 = fig1.add_subplot(3, 1, 1)
ax2 = fig1.add_subplot(3, 1, 2)
ax3 = fig1.add_subplot(3, 1, 3)

fig1
```

### 약간의 응용

```python
fig2 = plt.figure(figsize=(7, 7), facecolor='gray')
```

```python
ax1 = fig2.add_subplot(2, 2, 1)
ax2 = fig2.add_subplot(2, 2, 2)
ax3 = fig2.add_subplot(2, 1, 2)

fig2
```

```python
fig3 = plt.figure(figsize=(7, 7), facecolor='gray')
```

```python
ax1 = fig3.add_subplot(2, 3, 1)
ax2 = fig3.add_subplot(2, 3, 2)
ax3 = fig3.add_subplot(1, 3, 3)
ax4 = fig3.add_subplot(2, 3, 4)
ax5 = fig3.add_subplot(2, 3, 5)

fig3
```

## 3. plt.subplots()
- plt.subplots() = plt.figure() + fig.add_subplot()
- ax들을 규칙적으로는 그릴 수 있지만 위의 케이스와 같이 불규칙적으로는 그릴 수 없다.

```python
# 배열을 언팩할 수 있다.
# fig, (ax1, ax2) = plt.subplots(2, 1) 와 동일한 코드
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(7, 7), facecolor='orange')
```

```python
print(type(axes))
print(axes[0])
print(axes[1])
```

```python
# 2차원
fig, axes = plt.subplots(2, 2, figsize=(7, 7), facecolor='orange')

print(axes.shape)
print(axes)

for ax in axes.flat:
    print(ax)
    
for idx, ax in enumerate(axes.flat):
    print(idx, ax)
```

## 4. plt.subplot2grid
- 복잡한 형태의 그래프 레이아웃을 구성할 수 있다.
- fig.add_subplot(), plt.subplots()로는 구현하기 힘든 레이아웃을 만들 수 있다.

```python
fig = plt.figure(figsize=(7, 7), facecolor='lightgreen')

ax1 = plt.subplot2grid((2, 1), (0, 0), fig=fig)
ax2 = plt.subplot2grid((2, 1), (1, 0), fig=fig)
```

```python
# 2차원
fig = plt.figure(figsize=(7, 7), facecolor='lightgreen')

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2, fig=fig)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=3, fig=fig)
ax3 = plt.subplot2grid((3, 3), (2, 0), fig=fig)
ax4 = plt.subplot2grid((3, 3), (2, 1), fig=fig)
ax5 = plt.subplot2grid((3, 3), (2, 2), fig=fig)
```

```python
# fig.add_subplot(), plt.subplot2grid() 메서드를 사용해서 동일한 결과를 얻을 수 있다.
fig = plt.figure(figsize=(7, 7), facecolor='gray')
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 2, 3)
ax3 = fig.add_subplot(2, 2, 4)

fig2 = plt.figure(figsize=(7, 7), facecolor='lightgreen')
ax4 = plt.subplot2grid((2, 2), (0, 0), colspan=2, fig=fig2)
ax5 = plt.subplot2grid((2, 2), (1, 0), fig=fig2)
ax6 = plt.subplot2grid((2, 2), (1, 1), fig=fig2)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='lightgreen')
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2, fig=fig)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2, fig=fig)
ax3 = plt.subplot2grid((3, 3), (2, 0), colspan=2, fig=fig)
ax4 = plt.subplot2grid((3, 3), (0, 2), rowspan=3, fig=fig)
```

```python
fig = plt.figure(figsize=(7, 7), facecolor='lightgreen')

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3, fig=fig)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2, fig=fig)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2, fig=fig)
ax4 = plt.subplot2grid((3, 3), (2, 0), fig=fig)
ax5 = plt.subplot2grid((3, 3), (2, 1), fig=fig)
```
