# ➰ Spines
- Spine은 ax의 변을 의미한다. 단, tick과 ticklabel과는 관련이 없다.
- Spine 객체는 OrderedDict → keys, values, items 메서드 사용이 가능
## 목차
1. spine.set_visible
2. spine.set_position

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    print(spine_loc, spine, sep=' - ')
    
for spine_loc, spine in ax.spines.items():
    if spine_loc in ['left', 'bottom']:
        pass
    if spine_loc in ['right', 'top']:
        pass
```

## 1. spine.set_visible
False 인수를 전달하면 표시 X

```python
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['top', 'right']:
        spine.set_visible(False)
```

```python
# tick, ticklabel과는 별개
fig, ax = plt.subplots(figsize=(5, 5))

ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.tick_params(left=False, labelbottom=False)
```

```python
# spine과 tick 함께 사용
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['left', 'bottom']:
        spine.set_visible(False)
        
ax.tick_params(left=False, labelleft=False,
               right=True, labelright=True,
               bottom=False, labelbottom=False,
               top=True, labeltop=True)
```

```python
# 이미지 그리기
np.random.seed(0)

noise_img = np.random.normal(0, 1, (100, 100))

fig, ax = plt.subplots(figsize=(5, 5))
ax.imshow(noise_img)

for spine_loc, spine in ax.spines.items():
    spine.set_visible(False)
    
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

## 2. spine.set_position
- 2가지 매개변수: axes, data
    - 🔸 axes보다 data를 사용!

```python
# set_position('center') → LITERALLY spine을 중앙으로 이동시킴 (단점)
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['right', 'top']:
        spine.set_visible(False)
        
    if spine_loc in ['left', 'bottom']:
        spine.set_position('center')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['bottom', 'right', 'top']:
        spine.set_visible(False)
        
ax.tick_params(bottom=False, labelbottom=False)

# ax.spines['left'].set_position('center')과 동일
ax.spines['left'].set_position(('axes', 0.5))
```

```python
x = np.random.normal(2, 3, (100, ))
y = np.random.normal(2, 3, (100, ))

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(x, y)

ax.tick_params(labelsize=15)
fig.tight_layout()

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['right', 'top']:
        spine.set_visible(False)
        
    if spine_loc in ['left', 'bottom']:
        spine.set_position(('data', 0))
```

```python
x = np.random.normal(5, 3, (500, ))
y = np.random.normal(3, 3, (500, ))

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x, y)
ax.scatter(5.0, 3.0, color='orange')

ax.tick_params(labelsize=15, colors='red')
fig.tight_layout()

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['right', 'top']:
        spine.set_visible(False)
    
    if spine_loc == 'left':
        spine.set_position(('data', 5))
        spine.set_color('red')
        
    if spine_loc == 'bottom':
        spine.set_position(('data', 3))
        spine.set_color('red')
```
