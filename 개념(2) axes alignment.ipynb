# axes alignment

```python
import matplotlib.pyplot as plt
```

## ✅ fig.tight_layout
- 조정 대상 = title + xlabel + ylabel
- 매개변수
    - pad (간격 설정; ax 사이의 간격과 fig의 상하좌우 간격을 모두 설정)

```python
fig, ax = plt.subplots(figsize=(5, 5), facecolor='lightcoral')

ax.set_title('This is title', fontsize=15)
ax.set_xlabel('X Label', fontsize=10)
ax.set_ylabel('Y Label', fontsize=10)
```

```python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['label_x' + str(i) for i in range(4)]
ylabel_list = ['label_y' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(7, 7), facecolor='lightcoral')
for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx], fontsize=20)
    ax.set_xlabel(xlabel_list[ax_idx], fontsize=10)
    ax.set_ylabel(ylabel_list[ax_idx], fontsize=10)
fig.tight_layout(pad=2)
```

### (참고) axis.set_visible
- axis 눈금을 없애기 때문에 이미지를 그릴 때 유용하다.

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10), facecolor='moccasin')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```

## ✅ fig.subplots_adjust
- ax 사이의 간격과 fig의 상하좌우 간격을 각각(respectively) 설정할 수 있다.
- 6개의 매개변수
    - positional args
        - : left, right, bottom, top
        - fig의 상하좌우 간격을 조정한다.
        - As a whole, axes들이 차지하는 위치를 설정할 수 있다.
    - spacing args
        - : wspace, hspace
        - ax 사이의 간격을 조정한다.

### positional args

```python
fig, axes = plt.subplots(2, 2, figsize=(5, 5), facecolor='skyblue')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom=0.01, top=0.99)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(5, 5), facecolor='skyblue')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(left=0.01, right=0.99)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(5, 5), facecolor='skyblue')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom=0.01, top=0.99, left=0.01, right=0.99)
```

### spacing args

```python
fig, axes = plt.subplots(2, 2, figsize=(5, 5), facecolor='skyblue')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(hspace=0.05)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(5, 5), facecolor='skyblue')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(wspace=0.05)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(7, 7), facecolor='skyblue')

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

fig.subplots_adjust(hspace=0.2, wspace=0.05)
```

### 예제

```python
import numpy as np
fig = plt.figure(figsize=(12, 13), facecolor='skyblue')

# axes 그룹1
axes_g1 = np.empty(shape=(0, ))
main_g1 = plt.subplot2grid((5, 4), (0, 0), 2, 2, fig=fig)
main_g1.get_xaxis().set_visible(False)
main_g1.get_yaxis().set_visible(False)
axes_g1 = np.append(axes_g1, main_g1)

for r_idx in range(2, 5):
    for c_idx in range(2):
        ax = plt.subplot2grid((5, 4), (r_idx, c_idx), fig=fig)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        axes_g1 = np.append(axes_g1, ax)
axes_g1 = axes_g1.reshape(1, -1)        

# axes 그룹2
axes_g2 = np.empty(shape=(0, ))
main_g2 = plt.subplot2grid((5, 4), (0, 2), 2, 2, fig=fig)
main_g2.get_xaxis().set_visible(False)
main_g2.get_yaxis().set_visible(False)
axes_g2 = np.append(axes_g2, main_g2)

for r_idx in range(2, 5):
    for c_idx in range(2, 4):
        ax = plt.subplot2grid((5, 4), (r_idx, c_idx), fig=fig)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        axes_g2 = np.append(axes_g2, ax)
axes_g2 = axes_g2.reshape(1, -1)

fig.subplots_adjust(bottom=0.03, top=0.97, left=0.03, right=0.97, hspace=0, wspace=0)
```
