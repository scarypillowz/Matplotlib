## 1) 매개변수

### autopct

```python
np.random.seed(0)

n_can = 5
votes = np.random.randint(100, 1000, (n_can,))
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_votes = np.sum(votes)
percentages = votes / total_votes * 100

fig, ax = plt.subplots(figsize=(7, 7),
                       facecolor='lightsteelblue')
textprops = {'fontsize': 15, 'color': 'w'}
ax.pie(percentages,
       autopct='%1.1f%%',
       textprops=textprops)
```

### startangle, textprops
- startangle: pie plot 시작 지점
- textprops: 텍스트 커스터마이징

```python
# 정렬
np.random.seed(0)

n_can = 5
votes = np.random.randint(100, 1000, (n_can,))
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_votes = np.sum(votes)
percentages = votes / total_votes * 100

sort_idx = np.argsort(percentages)
percentages_sort = percentages[sort_idx]
cans_sort = cans[sort_idx]

fig, ax = plt.subplots(figsize=(7, 7),
                       facecolor='lightsteelblue')
textprops = {'fontsize': 15, 'color': 'w'}
ax.pie(percentages_sort,
       autopct='%1.1f%%',
       textprops=textprops,
       startangle=90)
```

### wedgeprops, colors
- wedgeprops: 선 커스터마이징(배경과 선의 색이 같으면 세련된 디자인 가능)
- colors: 색 커스터마이징

```python
np.random.seed(0)

n_can = 5
votes = np.random.randint(100, 1000, (n_can,))
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_votes = np.sum(votes)
percentages = votes / total_votes * 100

sort_idx = np.argsort(percentages)
percentages_sort = percentages[sort_idx]
cans_sort = cans[sort_idx]

fig, ax = plt.subplots(figsize=(7, 7),
                       facecolor='lightsteelblue')
textprops = {'fontsize': 15, 'color': 'w'}
wedgeprops = {'edgecolor': 'lightsteelblue', 'linewidth': 2}
ax.pie(percentages_sort,
       autopct='%1.1f%%',
       startangle=90,
       textprops=textprops,
       wedgeprops=wedgeprops)
```

```python
np.random.seed(0)

n_can = 5
votes = np.random.randint(100, 1000, (n_can,))
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_votes = np.sum(votes)
percentages = votes / total_votes * 100

sort_idx = np.argsort(percentages)
percentages_sort = percentages[sort_idx]
cans_sort = cans[sort_idx]

textprops = {'fontsize': 15, 'color': 'w'}
wedgeprops = {'edgecolor': 'lightsteelblue', 'linewidth': 2}

# cmap 사용
cmap = plt.get_cmap('tab10')
colors = [cmap(i) for i in range(n_can)]

fig, ax = plt.subplots(figsize=(7, 7),
                       facecolor='lightsteelblue')
ax.pie(percentages_sort,
       autopct='%1.1f%%',
       startangle=90,
       textprops=textprops,
       wedgeprops=wedgeprops,
       colors=colors)
```

### explode
강조 기능

```python
np.random.seed(0)

n_can = 5
votes = np.random.randint(100, 1000, (n_can,))
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_votes = np.sum(votes)
percentages = votes / total_votes * 100

sort_idx = np.argsort(percentages)
percentages_sort = percentages[sort_idx]
cans_sort = cans[sort_idx]

textprops = {'fontsize': 15, 'color': 'w'}
wedgeprops = {'edgecolor': 'lightsteelblue', 'linewidth': 2}

cmap = plt.get_cmap('tab10')
colors = [cmap(i) for i in range(n_can)]

fig, ax = plt.subplots(figsize=(7, 7),
                       facecolor='lightsteelblue')
explode = [0, 0, 0, 0, .1]
ax.pie(percentages_sort,
       autopct='%1.1f%%',
       startangle=90,
       textprops=textprops,
       wedgeprops=wedgeprops,
       colors=colors,
       explode=explode)
```

## 2) return values of pie plot

```python
np.random.seed(0)

n_can = 5
votes = np.random.randint(100, 1000, (n_can,))
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_votes = np.sum(votes)
percentages = votes / total_votes * 100

sort_idx = np.argsort(percentages)
percentages_sort = percentages[sort_idx]
cans_sort = cans[sort_idx]

textprops = {'fontsize': 15, 'color': 'w'}
wedgeprops = {'edgecolor': 'lightsteelblue', 'linewidth': 2}

cmap = plt.get_cmap('tab10')
colors = [cmap(i) for i in range(n_can)]

fig, ax = plt.subplots(figsize=(7, 7),
                       facecolor='lightsteelblue')
explode = [0, 0, 0, 0, .1]
wedges, texts, autotexts = ax.pie(percentages_sort,
                                  autopct='%1.1f%%',
                                  startangle=90,
                                  textprops=textprops,
                                  wedgeprops=wedgeprops,
                                  colors=colors,
                                  explode=explode)
# 큰 wedge부터 legend에 표시
# label은 cans_sort[::-1]
ax.legend(wedges[::-1], cans_sort[::-1],
          title='Candidates',
          title_fontsize=15,
          fontsize=15,
          facecolor='None',
          edgecolor='None',
          loc='center left',
          bbox_to_anchor=(.9, .5))
fig.tight_layout()
```

## 3) multiple pie plots
- 매개변수 wedgeprops의 속성
    - width (도넛 넓이 조절)
- ax.pie 매개변수
    - pctdistance (텍스트 위치 조절)

```python
np.random.seed(0)

n_can = 5
votes = np.random.randint(100, 1000, (n_can,))
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_votes = np.sum(votes)
percentages = votes / total_votes * 100

sort_idx = np.argsort(percentages)
percentages_sort = percentages[sort_idx]
cans_sort = cans[sort_idx]

textprops = {'fontsize': 10, 'fontweight': 'semibold', 'color': 'w'}
wedgeprops = {'width': .2,
              'edgecolor': 'lightsteelblue',
              'linewidth': 2}

fig, ax = plt.subplots(figsize=(7, 7),
                       facecolor='lightsteelblue')
wedges, texts, autotexts = ax.pie(percentages_sort,
                                  autopct='%1.1f%%',
                                  startangle=90,
                                  textprops=textprops,
                                  wedgeprops=wedgeprops,
                                  pctdistance=.9)
fig.tight_layout()
```

```python
n_can = 5
cans = np.array(['A', 'B', 'C', 'D', 'E'])
votes = np.random.randint(100, 1000, (2, n_can))
votes_sum = np.sum(votes, axis=0)

total_vote = np.sum(votes_sum)
percentages = votes_sum / total_vote * 100
percentages_tmp = votes / total_vote * 100

# color 커스터마이징
cmap = plt.get_cmap('tab20c', lut=20)
outer_colors = [cmap(4*i) for i in range(5)]
inner_colors = [cmap(4*i+j+1) for i in range(5) for j in range(2)]

# plotting 
fig, ax = plt.subplots(figsize=(10, 10),
                       facecolor='lightsteelblue')
textprops = {'fontsize': 20, 'color': 'w'}
wedgeprops1 = {'width': .4,
              'edgecolor': 'lightsteelblue',
              'linewidth': 2}
wedgeprops2 = {'width': .3,
               'edgecolor': 'lightsteelblue',
               'linewidth': 2}
wedges, texts, autotexts = ax.pie(percentages,
                                  colors=outer_colors,
                                  textprops=textprops,
                                  wedgeprops=wedgeprops1,
                                  pctdistance=.8,
                                  autopct='%1.1f%%')
ax.pie(percentages_tmp.flatten(order='F'),
       colors=inner_colors,
       wedgeprops=wedgeprops2,
       radius=.55)
ax.legend(wedges[::-1], cans_sort[::-1],
          title='Candidate',
          title_fontsize=15,
          fontsize=15,
          loc='center left',
          bbox_to_anchor=(.9, .5),
          facecolor='lightsteelblue',
          edgecolor='None')
fig.tight_layout()
```
