## 1. title

### ‚úÖ figure title
- ÏûêÏ£º ÏÇ¨Ïö© X
- tight_layout() Î©îÏÑúÎìú Ï†ÅÏö© X
- ÍπîÎÅîÌïú Î†àÏù¥ÏïÑÏõÉ Íµ¨ÏÑ±ÏùÑ ÏúÑÌï¥ fig.subplots_adjust Î©îÏÑúÎìú ÏÇ¨Ïö©(recommended)

```python
fig, ax = plt.subplots(figsize=(5, 5), facecolor='lightcoral')
fig.suptitle('Title of a Figure', fontsize=20, fontfamily='monospace');
```

```python
ax_titles = ['title' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(5, 5), facecolor='lightcoral')
fig.suptitle('Title of a Figure', fontsize=20, fontfamily='monospace')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(ax_titles[ax_idx], fontsize=15, fontfamily='serif')
    
fig.subplots_adjust(bottom=0.05, top=0.8, hspace=0.3)
```

## 3. Ïó∞ÏäµÎ¨∏Ï†ú

```python
fig, ax = plt.subplots(figsize=(7, 7), facecolor='lightcoral')

ax_impulse = fig.add_axes([.2, .55, .25, .25])
ax_prob = fig.add_axes([.6, .55, .25, .25])

ax_impulse.get_xaxis().set_visible(False)
ax_impulse.get_yaxis().set_visible(False)
ax_prob.get_xaxis().set_visible(False)
ax_prob.get_yaxis().set_visible(False)

ax.set_title('Gaussian colored noise', fontsize=20)
ax.set_xlabel('time (s)', fontsize=15)
ax.set_ylabel('current (nA)', fontsize=15)

ax_impulse.set_title('Impulse response', fontsize=10)
ax_prob.set_title('Probability', fontsize=10);
```

## 4. text alignment
- ax.text Î©îÏÑúÎìú
- Îëê Ï¢ÖÎ•òÏùò alignment
    1. horizontal(ha)
    2. vertical(va)
- Í∏∞Ï§ÄÏ†ê ÏÑ§Ï†ïÏù¥ ÌïµÏã¨!
    - ÎîîÌè¥Ìä∏Îäî left, bottom
- legendÏóêÎèÑ Ï†ÅÏö©Ïù¥ Í∞ÄÎä•Ìïú Î≤îÏö©ÏÑ±Ïù¥ ÎÑìÏùÄ ÌäπÏÑ±

### ‚úÖ horizontal

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

ax.grid()
ax.tick_params(axis='both', labelsize=10)

# ax.text(x=0, y=0, s='Hello', va='center', ha='left', fontsize=30)
# ax.text(x=0, y=0, s='Hello', va='center', ha='center', fontsize=30)
ax.text(x=0, y=0, s='Hello', va='center', ha='right', fontsize=30);
```

### ‚úÖ vertical

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

ax.grid()
ax.tick_params(axis='both', labelsize=10)

# ax.text(x=0, y=0, s='Hello', va='top', ha='center', fontsize=30)
ax.text(x=0, y=0, s='Hello', va='bottom', ha='center', fontsize=30);
```

## 5. title alignment
- ax.set_title(..., y)
    - üî∏ y ÏùåÏàò Í∞ÄÎä•

```python
ax_titles = ['(a) CE - clean', '(b) CE - noisy', '(c) LSR - noisy', '(d) SL - noisy']

title_bottom = .9
fig, axes = plt.subplots(2, 2, figsize=(10, 10), facecolor='lightcoral')

fig.suptitle('Symmetric Cross Entropy', fontsize=25, y=title_bottom + .03, va='bottom')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(ax_titles[ax_idx], fontsize=15, y=-0.25)
    ax.set_xlabel('Number of Epochs', fontsize=10)
    ax.set_ylabel('Class-wise Test Accuracy', fontsize=10)

fig.subplots_adjust(top=title_bottom, bottom=.1, hspace=.3, wspace=.3)
```

## 6. text properties
- üî∏ Îß§Í∞úÎ≥ÄÏàò fontdict
    - fontsize, fontfamily, color, alpha Îì±
    - ÏΩîÎìúÏùò Ïú†ÏßÄÎ≥¥ÏàòÏÑ± Ìñ•ÏÉÅ
- ÌÖçÏä§Ìä∏ Í¥ÄÎ†® API

    - ax.set_xlabel, ax.set_ylabel

### ‚úÖ ÏòàÏ†ú 1

```python
titles = ['Training Image', '2 scales', '4 scales', '5 scales', '6 scales', '8 scales']
title_font_dict = {
    'fontsize': 20,
    'fontweight': 'bold'
}

fig, axes = plt.subplots(2, 3, figsize=(20, 10))

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(titles[ax_idx], fontdict=title_font_dict)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.tight_layout()
```

### ‚úÖ ÏòàÏ†ú 2

```python
titles = ['Blurry Image', 'Ours', 'Mohan et al.']

fig = plt.figure(figsize=(24, 10))

axes = np.empty(shape=(0, 4))
n_group, n_col, n_row = 3, 4, 5

for g_idx in range(n_group):
    axes_row = np.empty(shape=(0, ))
    
    axes_row = np.append(axes_row, plt.subplot2grid((5, 12), (0, g_idx*n_col), colspan=4, rowspan=3, fig=fig))
    axes_row = np.append(axes_row, plt.subplot2grid((5, 12), (3, g_idx*n_col), colspan=3, rowspan=2, fig=fig))
    axes_row = np.append(axes_row, plt.subplot2grid((5, 12), (3, 3 + g_idx*n_col), fig=fig))
    axes_row = np.append(axes_row, plt.subplot2grid((5, 12), (4, 3 + g_idx*n_col), fig=fig))
    for ax in axes_row:
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    axes = np.vstack([axes, axes_row])
    axes_row[0].set_title(titles[g_idx], fontsize=30, y=-.8)

fig.subplots_adjust(bottom=.1, top=.95, left=.05, right=.95, hspace=.05, wspace=.05)
```
